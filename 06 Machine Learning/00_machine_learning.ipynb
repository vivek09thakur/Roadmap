## **Machine Learning**
> The topic which doesn't needs any intro! Let's dive into actual world of **Machine Learning** to see if the machines can actually learn?

**Do Machines Learn can actually learn?**

The idea of machine learning is to develop a model that can learn from data and make predictions on data. The model is trained on a set of training data and evaluated on a set of test data. The goal is to develop a model that generalizes well on unseen data.

---

1. **What is Machine Learning?**

> Machine Learning is a field of study that gives computers the ability to learn without being explicitly programmed. It is a branch of artificial intelligence based on the idea that systems can learn from data, identify patterns and make decisions with minimal human intervention.

2. **Machine Learning vs Artificial Intelligence**

> **Artificial Intelligence** is a broader concept that refers to machines being able to carry out tasks in a way that we would consider “smart”. **Machine Learning** is a current application of AI based around the idea that we should really just be able to give machines access to data and let them learn for themselves.

We can say that **Machine Learning** is a subset of **Artificial Intelligence**.

---


### **What is the need of Machine Learning?**

Machine Learning makes it possible to quickly and automatically produce models that can analyze bigger, more complex data and deliver faster, more accurate results – even on a very large scale. And by building precise models, an organization has a better chance of identifying profitable opportunities – or avoiding unknown risks.

### **Advantages of Machine Learning**

- **Handling multi-dimensional and multi-variety data:** Machine Learning algorithms can handle multi-dimensional and multi-variety data. Traditional computing algorithms are not capable of handling such data.

- **Handling missing and incomplete data:** Machine Learning algorithms can handle missing and incomplete data. Traditional computing algorithms are not capable of handling such data.

- **Handling changing in data:** Machine Learning algorithms can handle changing in data. Traditional computing algorithms are not capable of handling such data.

- **Handling large amount of data:** Machine Learning algorithms can handle large amount of data. Traditional computing algorithms are not capable of handling such data.

- **Handling non-linear data:** Machine Learning algorithms can handle non-linear data. Traditional computing algorithms are not capable of handling such data.

- **Handling noisy data:** Machine Learning algorithms can handle noisy data. Traditional computing algorithms are not capable of handling such data.

etc.


### **What are Machine Learning Algorithms?**
> Machine Learning algorithms are programs that can automatically learn from data without being explicitly programmed. Machine Learning algorithms are of two types: **Supervised Learning** and **Unsupervised Learning**.


### **What are Machine Learning Models?**
> Machine Learning models are programs that can automatically learn from data without being explicitly programmed. When we say a model ,Then the word "model" is typically used to describe a process for generating data, similar to the dataset at hand.

Therefore, good models can also be thought of as simplified versions of the real (unknown) data-generating process, capturing aspects that are relevant for modeling the data and extracting
hidden patterns from it.

- **What do you mean by ML Model?**

> To achieve this goal, we design models that are typically related to the process that generates data, similar to model the dataset we are given. 

- **Do a model learns? Or What do you understand when we say a model learned**

> A model is said to learn from data if its performance on a given task improves after the data is taken into account.
