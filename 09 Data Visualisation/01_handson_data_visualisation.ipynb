## **HandsOn Data Visualisation**
In this notebook, we will learn how to use the Python library `matplotlib` to create visualisations of data. We will also learn how to use the `pandas` library to read data from a CSV file and create a `DataFrame` object. We will then use the `DataFrame` object to create visualisations of the data.

1. **Creating a simple plot**

```python
# create a dataframe from the csv file
import pandas as pd

df = pd.read_csv('/Roadmap/sample data/businesses.csv')

# display the first 5 rows of the dataframe
print(df.head())
```

```python
import matplotlib.pyplot as plt

# create a simple plot of the number of business and founded year
# select only 5 rows of data
businesses = df['business']
most_top_businesses = businesses.head()

year_founded = df['year_founded']
year_founded_top = year_founded.head()

plt.plot(most_top_businesses, year_founded_top)
plt.show()
```

```python
# Customising the plot
plt.title('Number of businesses and year founded')
plt.xlabel('Number of businesses')
plt.ylabel('Year founded')
plt.plot(most_top_businesses, year_founded_top, marker='o', color='red', linestyle='--', linewidth=2, markersize=12)
plt.show()
```
