## **Different types of charts**

- **Line chart**

![Line chart](https://upload.wikimedia.org/wikipedia/commons/thumb/b/bd/Pushkin_population_history.svg/1200px-Pushkin_population_history.svg.png
)
Line charts are used to display trends over time. Use a line chart if you have text labels, dates or a few numeric labels on the horizontal axis.

> **Area charts** are just line charts, but with the area below the line filled in with a color or pattern. They are used to display accumulated totals over time.

- **Bar chart**

![Pie Chart](https://datavizproject.com/wp-content/uploads/types/Bar-Chart-Vertical.png)

Bar charts are used to compare different categories and show the frequency of the different categories. Use a bar chart if you have large text labels.
