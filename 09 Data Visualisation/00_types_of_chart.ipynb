## **Different types of charts**

- **Line chart**

![Line chart](https://upload.wikimedia.org/wikipedia/commons/thumb/b/bd/Pushkin_population_history.svg/1200px-Pushkin_population_history.svg.png
)
Line charts are used to display trends over time. Use a line chart if you have text labels, dates or a few numeric labels on the horizontal axis.

> **Area charts** are just line charts, but with the area below the line filled in with a color or pattern. They are used to display accumulated totals over time.

- **Bar chart**

![Pie Chart](https://datavizproject.com/wp-content/uploads/types/Bar-Chart-Vertical.png)

Bar charts are used to compare different categories and show the frequency of the different categories. Use a bar chart if you have large text labels.

The **horizontal bar chart** is a variation on the standard bar chart where the bars are displayed horizontally.


- **Pie chart**

![Pie Chart](https://datavizproject.com/wp-content/uploads/types/Pie-Chart.png)

Pie Charts are used to show parts to the whole, and often a % share. Use a pie chart when you want to compare parts of a single data point to the whole.



**Scatter plot**

![Scatter plot](https://www.w3schools.com/python/img_scatterplot.png)

Scatter plots are used to display the relationship between two continuous variables x and y. Use a scatter plot if you have two columns of data that you want to correlate.

**Histogram**

![Histogram](https://www.investopedia.com/thmb/ilebTtfvaQER39ue8gEkMIpP6MY=/1500x0/filters:no_upscale():max_bytes(150000):strip_icc()/Histogram1-92513160f945482e95c1afc81cb5901e.png)

Histograms are used to show the distribution of a numerical variable. Use a histogram when you have a continuous numerical data and want to understand the distribution.


**Box and Whisker plot**

![Box and Whisker plot](https://www.investopedia.com/thmb/ilebTtfvaQER39ue8gEkMIpP6MY=/1500x0/filters:no_upscale():max_bytes(150000):strip_icc()/Histogram1-92513160f945482e95c1afc81cb5901e.png))

Box and Whisker plots are used to show the distribution of categorical data. Use a box and whisker plot when you have multiple data sets from independent sources that are related in some way.
