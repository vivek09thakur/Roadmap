**Making a model to predict the stock price of TATA MOTORS using linear regression**
> Here we only use the closing price of the stock to predict the price of the stock for the next day

```python
import pandas as pd

df = pd.read_csv('/Projects/LinearRegression_Stock_Prediction/dataset/tesla_data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['Open', 'High', 'Low']], df['Adj Close'], test_size=0.2,
                                                    random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
print(f'Predicted values :\n {y_pred}')
```

```python
# Reshape the data to same size
y_pred = y_pred.reshape(len(y_pred), 1)
y_test = y_test.values.reshape(len(y_test), 1)
```

```python
# visualize the results
fig_2 = plt.figure(figsize=(40, 15))
plt.plot(y_test, color='blue', label='Actual Tesla Stock Price')
plt.plot(y_pred, color='red', label='Predicted Tesla Stock Price')
plt.title('Tesla Stock Price Prediction')
plt.xlabel('Date')  
plt.ylabel('Tesla Stock Price')
plt.legend()
plt.show()
```
