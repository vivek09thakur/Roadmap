## KNN CLASSIFIER OR K-NEAREST NEIGHBORS CLASSIFIER
> KNN is one of the fundamental supervised machine learning algorithms mostly used for classification problems. It is a non-parametric and lazy learning algorithm. Non-parametric means there is no assumption for underlying data distribution. In other words, the model structure determined from the dataset

```python
# Dataset: movie_dataset.csv
movie_dataset = '/Roadmap/Sample Dataset/movie_dataset.csv'
```

```python
# Importing the dataset
import pandas as pd

movie_data = pd.read_csv(movie_dataset)
```
