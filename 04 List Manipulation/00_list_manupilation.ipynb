## LIST MANUPILATION

**What is a list in python?**

List are some compound data types in python. List can be used to store a collection of data. List are mutable, which means that we can change the value of the list after it has been created. List are ordered, which means that the items have a defined order, and that order will not change. List allows duplicate members.

**How to create a list?**
List are created by placing all the items (elements) inside a square bracket [ ], separated by commas. It can have any number of items and they may be of different types (integer, float, string etc.).

```python
# Creating a list
my_list = [1, 2, 3, 4, 5]
print(my_list)
```

````python
my_list = [1, 2, 3, 4, 5]
print(my_list)
print('type(my_list):', type(my_list))
````
