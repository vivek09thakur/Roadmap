> In programming , data type describes what kind of data is stored in a variable. It also describes what operations can be performed on that variable.

Data Types is also divided into two catogeories :
- *Primitive data type*

    Primitive data types (pre-defined data type,pre-exsisting data types,core data types) are those data types which are built-in or pre exsist. It is not defined by user. Python have these data types as core data types :
    
    - Integer - `int`
    - String - `str`
    - Float - `float`
    - Boolean - `bool`
    - List - `list`
    - Dictionary - `dict`
    

- *Non-primitive data type*

    Non-primitive data type are the user defined data types which is defined by user itself. `array,link_list,tree,graph,stack` etc are the example of Non-primitive data type.

### 1. String

> The string data type is used to represent text data. It is a sequence of characters enclosed in single quotes, double quotes, or triple quotes. Strings are immutable in Python, which means that once a string is created, it cannot be modified. Instead, any operation that appears to modify a string actually creates a new string object.

````python
str_example = 'Hello Hie bye bye'
type(str_example)
````

**String Concatenation**
String concatenation is the process of joining two or more strings together to create a new string. In Python, we can concatenate strings with the + operator.

````python
# Here is an example of string concatenation
string_1 = 'Hello'
string_2 = 'World'

# First way
string_3 = string_1 + string_2
print(string_3)
````

````python
# Second way
print(string_1 + string_2)
````

````python
# Third way
print('Hello' + 'World')
````

### 2. Integer

All numeric values with no decimal values such as : `1,2,3,4,-1,-2,-3,0` are integers in python having `int` data type. Integers can be positive or negative *(If the `+` or `-` symbol is missing then it is considered as positive integer)*.

**Integers in python is classified into two categories :**

- **Integers (signed).** It is the normal integers or whole numbers : `1,2,3,4,5,-1,-2,-3,0` etc

````python
# Declaring a variable num
num = 69
print(num)
print(type(num))
````

- **Booleans** 
These represents the value of `true` or `False`. The boolean type is a subtype of plain integers, and Boolean values `False` or `True` behave like the values 0 and 1, repectively.

To get the boolean equivalent of `0` and `1` , we can use `bool(0)` and `bool(1)` for True and False.

````python
# Boolean equivalent to True
bool(1)
````

````python
# Boolean equivalent of False
bool(0)
````

### 3. Floating Point Numbers

A number having fractional part is a floating point number. For example , 3.14159 is a floating point number. The decimal point describes that it is a floating point number. For example , `12` and `12.0` in mathematics they both have same value but in python 3 they can be considered as two different varaible with two different data types.

````python
num1 = 12
type(num1)
````

````python

num2 = 12.0
type(num2)
````
