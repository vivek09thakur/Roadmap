- **Data Type**
> In programming , data type describes what kind of data is stored in a variable. It also describes what operations can be performed on that variable.

Data Types is also divided into two catogeories :
- *Primitive data type*

    Primitive data types (pre-defined data type,pre-exsisting data types,core data types) are those data types which are built-in or pre exsist. It is not defined by user. Python have these data types as core data types :
    
    - Integer - `int`
    - String - `str`
    - Float - `float`
    - Boolean - `bool`
    - List - `list`
    - Dictionary - `dict`
    

- *Non-primitive data type*

    Non-primitive data type are the user defined data types which is defined by user itself. `array,link_list,tree,graph,stack` etc are the example of Non-primitive data type.

But here we are only focusing on *Primitive data type*. Let's start one-by-one
