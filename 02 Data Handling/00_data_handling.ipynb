## Data Handling

```python
import matplotlib.pyplot as plt
import random

list_1 = []
list_2 = []
for i in range(69):
    list_1.append(random.randint(1,20))
    list_2.append(random.randint(3,90))

plt.plot(list_1,list_2,scaley=1)
plt.ylabel('Time')
plt.xlabel('My Mood Swings')
plt.title('Visual Representation of my mood')
plt.show()
    
```
