### 4. Tuples and Lists

Python provides two compound data types for grouping values together : **tuples** and **lists**.
> Compound data types are those data types which can store multiple values in it. It can be of any data type.

For example , `list` can store multiple values of any data type in it. It can store `int`,`str`,`float` etc in it. It can also store `list` in it. It can also store `tuple` in it. It can also store `dict` in it. It can also store `set` in it. It can also store `bool` in it.

**Tuples** and **Lists** are very similar to each other. The only difference between them is that tuples are immutable whereas lists are mutable.

Let's start with tuples.

**1. Tuples**

Tuples are immutable compound data type in python3. It means that once a tuple is created, we cannot modify it. Tuples are created by placing comma-separated values inside parentheses `()`. For example, `t = (1, 2, 3)` is a tuple containing three values.



````python
# Creating a tuple
tuple_1 = (1, 2, 3, 4, 5)
print(tuple_1)
print(type(tuple_1))
````

**Acessing Tuple Elements**

We can access the elements of a tuple using the index operator `[]`. The index of the first element is 0, the index of the second element is 1, and so on. For example, `t[0]` returns the first element, `t[1]` returns the second element, and so on.

````python
# Create tuple with mixed datatypes
tuple_2 = (1, 'Hello', 3.4)

# Lets see how to access each element of a tuple
print(tuple_2[0])
````

````python
# We can also use negative indexing
print(tuple_2[-1])
````

````python
# Slicing in tuple
print(tuple_2[0:2])
````

````python
# Adding a new element to a tuple
tuple_2 = tuple_2 + (5,)
print(tuple_2)
````

````python
# Removing an element from a tuple
tuple_2 = tuple_2[:-1]
print(tuple_2)
````

````python
# Another way to remove an element from a tuple
tuple_2 = tuple_2[0:2]
print(tuple_2)
````

> **Modifying Tuple Elements :** Since tuples are immutable, we cannot modify their elements. For example, `t[0] = 5` will raise an error.

**2. Lists**

List are the immutable compound data type in python3. It means that once a list is created, we can modify it. Lists are created by placing comma-separated values inside square brackets `[]`. For example, `l = [1, 2, 3]` is a list containing three values.

**Acessing List Elements**

````python
# Aceesing each element of a list
list_1 = [1, 2, 3, 4, 5]
print(list_1[0])
````

````python
# Adding a new element to a list
list_1.append(6)
print(list_1)
````

````python
# Removing an element from a list
list_1.remove(6)
print(list_1)
````

````python
# Counting the number of times an element is present in a list
list_1.count(1)
````

````python
# Deleting an element from a list
del list_1[0]
print(list_1)
````

````python
# Another way to delete an element from a list
list_1.pop(0)
print(list_1)
````

### 5. Dictionaries

It is also a compound data type in python3. It is also a mutable data type. It is also a key-value pair data type
Dictionaries are used to store key-value pairs. A dictionary is created by placing comma-separated key-value pairs inside curly braces `{}`. For example, `d = {'a': 1, 'b': 2, 'c': 3}` is a dictionary containing three key-value pairs.


````python
# Creating a dictionary
dict_1 = {'key1': 'value1', 'key2': 'value2'}
print(dict_1)
print(type(dict_1))
````

````python
# Accessing the value of a dictionary
print(dict_1['key1'])
````

````python
# Adding a new key-value pair to a dictionary
dict_1['key3'] = 'value3'
print(dict_1)
````

````python
# Deleting a key-value pair from a dictionary
del dict_1['key3']
print(dict_1)
````

````python
# Deleting a key-value pair from a dictionary using pop
dict_1.pop('key2')
print(dict_1)
````

### Immutable and Mutable Data Types

- **Immutable Data Types**

Immutable data types are those data types which cannot be modified once created. For example, integers, floating-point numbers, strings, tuples, and frozensets are immutable data types.

- **Mutable Data Types**

Mutable data types are those data types which can be modified once created. For example, lists, dictionaries, and sets are mutable data types.

**Immutable Data Types in Python**

Let us try to understand, how immutable data types works in python. Strings are immutable data types in python. Let's take an example.

````python
p = 5
q = p
r = 5

print('p = ',p)
print('q = ',q)
print('r = ',r)

# but here we can see that we can easily modified value of `a` and `b`. So how they are immutable data types?
````

````python
p = 10
r = 7
q = r

print('p = ',p)
print('q = ',q)
print('r = ',r)
````

**Python Variables are not storage containers**

Unlike other programming languages, python variables are not storage containers. They are just labels or names that are attached to the values.When we assign a value to a variable, we are actually binding the variable to the value. The value is stored in memory and the variable is just a label that points to it. And when we assign a new value to the variable, the old value is not overwritten. Instead, the variable is re-bound to the new value. The old value is discarded if it is not referenced by any other variable.

**Mutable Data Types in Python**

Mutable data types are those data types which can be modified once created. For example, lists, dictionaries, and sets are mutable data types.

### Operators in Python

In python, operators are special symbols that are used to perform operations on operands. For example, `+` is an operator that performs addition on two operands. Here, `a + b` is an expression that represents addition operation on two operands `a` and `b`.

**Types of Operators in Python**

Python supports the following types of operators.

- Arithmetic Operators
- Comparison Operators
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Identity Operators
- Membership Operators

Let's start with one-by-one.

**1. Arithmetic Operators**

Arithmetic operators are used to perform arithmetic operations on operands. For example, `+` is an arithmetic operator that performs addition on two operands. Here, `a + b` is an expression that represents addition operation on two operands `a` and `b`.

````python
# For example of arithmetic operators
a = 10
b = 5

# Addition of numbers
print('a + b =',a+b)
````

````python
# Subtraction of numbers
print('a - b =',a-b)
````

````python
# Division(float) of number

print('a / b =',a/b)
````

````python
# Multiplication of number
print('a * b =',a*b)
````

````python
# Modulo of both number

print('a % b =',a%b)
````

**2. Comparison Operators**

Comparison operators are used to compare two operands. For example, `==` is a comparison operator that compares two operands and returns `True` if they are equal, `False` otherwise. Here, `a == b` is an expression that represents comparison operation on two operands `a` and `b`.
