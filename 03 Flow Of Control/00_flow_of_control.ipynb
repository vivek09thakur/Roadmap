## FLOW OF CONTROL
In this chapter we are going to learn about how we can control the flow of our program. We will learn about conditional statements, loops and functions.


### Conditional Statements
Conditional statements are used to execute a block of code if a certain condition is true. If the condition is false, another block of code can be executed. The following conditional statements are available in Python:

- if statement: executes a block of code if a condition is true
- if...else statement: executes a block of code if a condition is true and another block of code if it is false
- if...elif...else statement: executes different blocks of code for different conditions

Let's look at some examples:

```python
## Example of If Statement
name = input("What is your name? ")
if name == "Vivek":
    print("Hello Vivek")
```

```python
999## Example of If-Else Statement
name = input("What is your name? ")
if name == "Vivek":
    print("Hello Vivek")
else:
    print("Hello Stranger")
```

```python
## Example of If-Elif-Else Statement
name = input("What is your name? ")
if name == "Vivek":
    print("Hello Vivek")
elif name == "Sam":
    print("Hello Sam")
elif name == "John":
    print("Hello John")
else:
    print("Hello Stranger")
```
