**Question 1**

```python
# simple interest = (principal x rate x time)/100
prin = float(input("Enter principal amount = "))
rate = float(input("Enter rate of interest = "))
time = float(input("Enter time period = "))

si = (prin*rate*time)/100
print("Simple Interest = ",si)
```

**Question 2**

```python
import numpy as np

temprature_list = []

for i in range(7):
    temp = float(input(f"Enter the temprature of day {i+1} = "))
    temprature_list.append(temp)
    
temp_array = np.array(temprature_list)
sum_temp = np.sum(temp_array)
avg_tem = sum_temp/7
print("Average temprature of the week = ",avg_tem)
```

**Question 3**

```python
import math 

x = float(input("Enter the value of x = "))
y = float(input("Enter the value of y = "))
z = float(input("Enter the value of z = "))

expression = 4*math.pow(x,4) + 3*math.pow(y,3) + 9*z + math.pi

print("4x^4 + 3y^3 + 9z + pi = ",expression)
```
