**4. Bitwise Operators**

Bitwise operators are used to perform bitwise operations on operands. For example, `&` is a bitwise operator that performs bitwise AND operation on two operands. Here, `a & b` is an expression that represents bitwise AND operation on two operands `a` and `b`.

There are six bitwise operators in python.

- `&` - Bitwise AND
- `|` - Bitwise OR
- `^` - Bitwise XOR
- `~` - Bitwise NOT
- `<<` - Bitwise Left Shift
- `>>` - Bitwise Right Shift


````python
# & Bitwise AND

# Bitwise AND works on bits and perform AND operation on each bit of the number. It returns 1 only if both bits are 1. Like normal AND operator, it operates on two numbers. It converts those numbers into binary and then do AND operation on each bit of both numbers.
a = True
b = False

print('a = ',a , 'and b = ',b)
print('a & b =',a & b)
print('b & b =',b & b)
print('a & a =',a & a)
print('b & a =',b & a)
````

````python
#  | Bitwise OR

# Bitwise OR works on bits and perform OR operation on each bit of the number. It returns 1 if either of the bit is 1 else returns 0. Like normal OR operator, it operates on two numbers. It converts those numbers into binary and then do OR operation on each bit of both numbers.

print('True or bitwise or True = ',True | True)
print('True or bitwise or False = ',True | False)
print('False or bitwise or False = ',False | False)
print('False or bitwise or True = ',False | True)
````

````python
# ^ Bitwise XOR
# Bitwise XOR works on bits and perform XOR operation on each bit of the number. It returns 1 if the corresponding bits are different else returns 0. Like normal XOR operator, it operates on two numbers. It converts those numbers into binary and then do XOR operation on each bit of both numbers.

# Here we have True ^ True = False, 
print('True or bitwise xor True = ',True ^ True)
# Here we have True ^ False = True,
print('True or bitwise xor False = ',True ^ False)
# False ^ False = False
print('False or bitwise xor False = ',False ^ False)
# False ^ True = True
print('False or bitwise xor True = ',False ^ True)
````

````python
# Bitwise NOT
# In Python, bitwise operators are used to perform bitwise calculations on integers. The integers are first converted into binary and then operations are performed on each bit or corresponding pair of bits, hence the name bitwise operators. The result is then returned in decimal format.

# Here we have not True = False
print('not True = ',~ True)
# Here we have not False = True
print('not False = ',~ False)
````

````python
# << Bitwise Left Shift

# Bitwise left shift operator shifts the bits of the number to the left and fills 0 on voids left as a result. Similar effect as of multiplying the number with some power of two.

# Here we have 2 << 1 = 4
print('2 << 1 = ',2 << 1)
# Here we have 2 << 2 = 8
print('2 << 2 = ',2 << 2)
# Here we have 2 << 3 = 16
print('2 << 3 = ',2 << 3)
````

````python
# >> Bitwise Right Shift

# Bitwise right shift operator shifts the bits of the number to the right and fills 0 on voids left as a result. Similar effect as of dividing the number with some power of two.

# Here we have 2 >> 1 = 1
print('2 >> 1 = ',2 >> 1)
# Here we have 2 >> 2 = 0
print('2 >> 2 = ',2 >> 2)
# Here we have 2 >> 3 = 0
print('2 >> 3 = ',2 >> 3)
````

**Operator Precedence**

Operator precedence determines the order in which operators are evaluated. Operators with higher precedence are evaluated first. For example, `2 + 3 * 5` is evaluated as `2 + (3 * 5)` because the `*` operator has higher precedence than the `+` operator.

**Operator Associativity**

Operator associativity determines the order in which operators with the same precedence are evaluated. Operators with left-to-right associativity are evaluated from left to right. For example, `2 - 3 + 5` is evaluated as `(2 - 3) + 5` because the `-` and `+` operators have left-to-right associativity. Operators with right-to-left associativity are evaluated from right to left. For example, `2 ** 3 ** 2` is evaluated as `2 ** (3 ** 2)` because the `**` operator has right-to-left associativity.

Let's understand it by an example.

````python
# Let's see how operators precedence works in python

# Here we have 2 + 3 * 5 = 17
print(2 + 3 * 5) # In this case multiplication will be performed first and then addition
````

````python
#  We can use parenthesis to change the order of precedence
print((2 + 3) * 5) # In this case addition will be performed first and then multiplication
````

````python
# Operator Associativity
# Associativity is used when two operators of same precedence appear in an expression. Associativity can be of two types: Left to Right or Right to Left. It depends on the order in which the expression is evaluated.

# Here we have 2 ** 3 ** 2 = 512
print(2 ** 3 ** 2) # In this case exponentiation will be performed from right to left
````

### Expressions in Python

An expression is a combination of values, variables, operators, and function calls that are evaluated to produce a value. For example, `2 + 3` is an expression that evaluates to `5`. Similarly, `x + y` is an expression that evaluates to `x + y`.

**Components of an Expression**

- **Values** - Values are the most basic components of an expression. For example, `2` and `3` are values in the expression `2 + 3`.
- **Variables** - Variables are used to store values. For example, `x` and `y` are variables in the expression `x + y`.
- **Operators** - Operators are used to perform operations on operands. For example, `+` is an operator that performs addition on two operands. Here, `2 + 3` is an expression that represents addition operation on two operands `2` and `3`.
- **Function Calls** - Function calls are used to call functions. For example, `print()` is a function call that calls the `print()` function. Here, `print(2 + 3)` is an expression that calls the `print()` function with the argument `2 + 3`.


````python
# Examples of expressions in python

print(2+3*5)
print((2+3)*5)
print(2**3**2)
print(2**9)
````

**Type Casting**

Type casting is the process of converting the value of one data type to another data type. For example, `int(3.5)` converts the floating-point number `3.5` to the integer `3`. It is also known as type conversion or type coercion.

````python
a = 10
print(float(a))
````
