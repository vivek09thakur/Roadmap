## DICTIONARIES

In python 3x, **dictionaries** are the compound data types which is used to store data values in key:value pairs. The values in dictionaries can be of any data type and can be duplicated, whereas keys can’t be repeated and must be immutable. The keys in dictionaries are used to access the values. Dictionaries are written with curly brackets and have keys and values in it.

**Why use dictionaries?**

Dictionaries are used to store data values in key:value pairs. A dictionary is a collection which is unordered, changeable and indexed. In Python dictionaries are written with curly brackets, and they have keys and values.

**Properties of Dictionaries**

- Dictionaries are unordered
- Dictionaries are mutable
- Dictionaries are indexed
- Dictionaries are written with curly brackets
- Dictionaries are changeable
- Dictionaries are unordered, so the order that the keys are added doesn’t necessarily reflect what order they may be reported back.
- Dictionaries are mutable, meaning they can be changed. The values of a dictionary can be changed, and a new key/value pair can be added to an existing dictionary.


### Creating a Dictionary

We can create a dictionary by placing a comma-separated list of key: value pairs in curly braces {}. Let's see how to create a dictionary in Python. Look at the following example.

````python
## Creating a dictionary
my_dict_1 = {
    "key1": "value1",
    "key2": "value2",
    "key3": "value3"
}

print(my_dict_1)
print('Data type:', type(my_dict_1))
````

### Accessing Elements from Dictionary

We can use the key to access the value from the dictionary. Like from the above example, we can access the value of the key "key1" to get the value "value1" as follows.

````python
## Accessing a value
print(my_dict_1["key1"])
````

### Adding a new key-value pair to a dictionary

### Modifying a key-value pair in a dictionary

When we want to modify a key-value pair in a dictionary, we can do so by simply accessing it and assigning a new value to it. Let's see how to do it.

````python
## Modifying a value
print('my dict 1 before:', my_dict_1)
my_dict_1['key1'] = 'new_value1'
print('my dict 1 after:', my_dict_1)
````

## Removing a key-value pair from a dictionary

There are several ways to remove items from a dictionary.Here are the methods to remove a key-value pair from a dictionary :

- **pop()** : The pop() method removes the item with the specified key name.
- **popitem()** : The popitem() method removes the last inserted item (in versions before 3.7, a random item is removed instead).
- **del** : The del keyword removes the item with the specified key name.
- **clear()** : The clear() method empties the dictionary.


**1. Deleting a key-value pair using pop() method**

In this method we can delete a key-value pair from a dictionary using the pop() method. The pop() method removes the item with the specified key name. The syntax of the pop() method is as follows :

syntax : dict.pop(keyname, defaultvalue)

````python
## Deleting a key-value pair using pop()
print('my dict 1 before:', my_dict_1)
my_dict_1.pop('key1')
print('my dict 1 after:', my_dict_1)
````

**2. Deleting a key-value pair using popitem() method**

In this method we can delete a key-value pair from a dictionary using the popitem() method. The popitem() method removes the last inserted item (in versions before 3.7, a random item is removed instead). The syntax of the popitem() method is as follows :

syntax : `dict.popitem()`


````python
## deleting a key-value pair using popitem()
print('my dict 1 before:', my_dict_1)
my_dict_1.popitem()
print('my dict 1 after:', my_dict_1)
````

**3. Deleting a key-value pair using del keyword**

In this method we can delete a key-value pair from a dictionary using the del keyword. The del keyword removes the item with the specified key name. The syntax of the del keyword is as follows :

syntax : `del dict[key]`

````python
## deleting a key-value pair using del
print('my dict 1 before:', my_dict_1)
del my_dict_1['key2']
print('my dict 1 after:', my_dict_1)
````

**4. Deleting a key-value pair using clear() method**

In this method we can delete a key-value pair from a dictionary using the clear() method. The clear() method empties the dictionary. The syntax of the clear() method is as follows : `dict.clear()`


````python
## deleting all key-value pairs using clear()
print('my dict 1 before:', my_dict_1)
my_dict_1.clear()
print('my dict 1 after:', my_dict_1)
````
