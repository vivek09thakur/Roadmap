## What is Dataframe?
- Dataframe is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object.
- Dataframe is a main object in Pandas. It is used to represent data with rows and columns (tabular or excel spreadsheet like data).

### Creating a dataframe

```python
import pandas as pd
std_data = [
    (1,'vivek', 23, 'M'),
    (2,'rahul', 24, 'M'),
    (3,'sairaj', 23, 'M'),
    (4,'srijani', 23, 'F'),
]
df = pd.DataFrame(std_data)
print(df)
```

#### Creating an empty dataframe
- A basic DataFrame, which can be created is an Empty Dataframe.

```python
# an empty dataframe
import pandas as pd

df = pd.DataFrame()
print(df)
```

## **Operations on pandas dataframe**

```python
import pandas as pd

df = pd.read_csv('/Roadmap/sample data/student.csv')

print(df)
```

**Selecting a column in pandas dataframe**

```python
age = df['student_age']
print(age)
```

**Selecting a row in pandas dataframe**

```python
row_1 = df.iloc[0]
print(row_1)
```

**Selecting multiple rows in pandas dataframe**

```python
df.iloc[[0,3]]
```

```python
df.iloc[[0,3], [0,1]]
```

**Filtering in pandas dataframe**

```python
df[df['student_age'] > 20]
```

**Adding a new column to existing DataFrame in Pandas**

```python
df['phone_numbers'] = [234567, 345678, 456789, 567890, 678901, 789012, 890123, 901234, 123456, 234567,23456,345678,56789,56768,57686,45678,56456,456789,567890,567689]
```

```python
print(df)
```

**Updating existing column of a DataFrame in Pandas**

```python
df.loc[df.index == 0, 'student_name'] = 'vivek'
```
