### **Self-Learning Image Classification Model**
> **Using Clustering and K-Means Algorithm**

In this project, I've tried to make a ML Model that can learn by itself and catgoerise similar kind of images in groups or classes. The model classify images of similar kind into groups and if a new image is given to this model then it can tell which class/group does the new image belongs to.

```python
from skimage import io, transform
from skimage.feature import hog
from sklearn.cluster import KMeans
import os
import shutil
import numpy as np
```

#### **Building the Model**

```python
class IMAGE_CLASSIFICATION_MODEL:
    
    def __init__(self, resize_shape=(128, 128), orientations=8, pixels_per_cell=(16, 16), block_norm='L2-Hys', cells_per_block=(1, 1)) -> None:
        self.resize_shape = resize_shape
        self.orientations = orientations
        self.pixels_per_cell = pixels_per_cell
        self.block_norm = block_norm
        self.cells_per_block = cells_per_block
    
    def extract_image_features(self, image_path):
        image = io.imread(image_path)
        resized_image = transform.resize(image, self.resize_shape)
        fd, hog_image = hog(resized_image, orientations=self.orientations, pixels_per_cell=self.pixels_per_cell,
                            block_norm=self.block_norm, cells_per_block=self.cells_per_block, visualize=True, channel_axis=-1)
        return fd
    
    def learn_and_classify(self, features, cluster_numbers):
        kmeans = KMeans(n_clusters=cluster_numbers)
        X = np.vstack(features)
        cluster_labels = kmeans.fit_predict(X)
        return cluster_labels
    
    def create_directory(self, image_paths, cluster_labels, destination_path):
        for i in range(max(cluster_labels) + 1):
            cluster_dir = os.path.join(destination_path, f'group_{i}')
            os.makedirs(cluster_dir, exist_ok=True)
            
        for image_path, cluster_label in zip(image_paths, cluster_labels):
            cluster_path = os.path.join(destination_path, f'group_{cluster_label}')
            shutil.copy(image_path, cluster_path)
```

```python

image_classifier = IMAGE_CLASSIFICATION_MODEL()
features = [image_classifier.extract_image_features(image) for image in images]
image_labels = image_classifier.learn_and_classify(features,2)
```
