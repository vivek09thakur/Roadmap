### Variable and Assignments

> *Variables* are basically a container that stores some data item.

Yeah basically this is the traditional defintion of a variable in any programming language but in python things works a bit differently not completely but a little bit differently.

Let's see how things works ...

````python
# CREATING A VARIABLE
var_a = 70
````

It is that simple to create a variable in python. As python is a **dyanamically typed language** which means you don't need to define *data type* before creating a variable like other middle-level languages like **C++,JAVA** etc.

####  IMPORTANT - VARIABLES ARE NOT STROAGE CONTAINERS IN PYTHON

**PYTHON VARIABLES ARE NOT CREATED AS STORAGE CONTAINER LIKE OTHER PROGRAMMING LANGUAGES DO**

Python variables are named labels that refers to a value. *Or in other words they are just objects of class that refers to a memory location were a data item stored*.

````python
var_b = 5
print(type(var_b))
````

**See , it says `<class 'int'>` , which means that `var_b` is an object of `class int` which is refering to a memory location having a value of 5 stored in it.**

````python
# MULTIPLE ASSIGNMENTS ( to single value)
var_c = var_d = var_e = 10
print(var_c)
print(var_d)
print(var_e)
````

````python
# MULTIPLE ASSIGNMENTS ( to different values)
var_c , var_d , var_e = 10 , 20 , 30
print(var_c)
print(var_d)
print(var_e)
````
