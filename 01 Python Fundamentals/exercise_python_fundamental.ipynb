````python
# 3. write a program to obtain length and breadth of a rectangle and prints it area
length = float(input('enter the length of rectangle = '))
breadth = float(input('enter the breadth of rectangle = '))

area_of_rect = length*breadth
print('length = ',length)
print('breadth = ',breadth)
print('area of rectangle = ',area_of_rect,'m^2')
````

````python
# 4. write a program to calculate BMI of a person
weigth = float(input('enter your weigth in kg = '))
height = float(input('enter your height in metres = '))

bmi = weigth/(height*height)
print('bmi of person having weigth = ',weigth,'kg and height = ',height, 'metres is ',bmi)
````
